```python
import json
import pandas as pd
```

```python
def convert_json_to_dataframe(filename):
    file=open(filename)
    data=[json.loads(l) for l in file]
    dataframe=pd.json_normalize(data)
    return dataframe
```

```python
df_users=convert_json_to_dataframe('users.json')
```

```python
df_brand=convert_json_to_dataframe(r'brands.json')
```
